## Ndarray


```python
import numpy as np

#1d array

arr1 = np.array([1,2,3,4,5])
print(type(arr1))
print(arr1.ndim)
```

```python
#2d array

arr2 = np.array([[1,2,3], #row1
                  [4,5,6] #row2
                 ])

print(arr2.ndim)
```

```python
#3d array - akta array r moddhe onekgula 2d array

arr3 = np.array([

    #1st floor
    [[1,2,3],
     [4,5,6]],

    #2nd floor
    [[6,7,8],
     [9,10,11]],

    #3rd floor
    [[1,2,3],
     [4,5,6]],

    #4th floor
    [[6,7,8],
     [9,10,11]]
])

print(arr3.ndim)
```

## Attributes


```python
#dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#shape -> tuple (row,column)
print(arr1.shape)
print(arr2.shape)
print(arr3.shape) # (floor, row,column)


#data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

#size -> total koyta value ase
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

## Data Types

```python
arr = np.array([1,2,3])
print(arr.dtype)

arr = np.array([11.2,1.3,22.0])
print(arr.dtype)

arr = np.array(["sadman","tutu","kutu"])
print(arr.dtype) #U -> Unicode
```

Upcasting

```python
arr = np.array([1,2,3.5])
print(arr.dtype) # upcasted to float

arr = np.array([2.5,3,'3'])
print(arr.dtype) # upcasted to string
```

Selecting a datatype

```python
#using dtype
arr = np.array([1,2,3],dtype=np.int32)
print(arr.dtype)

#using astype

arr = np.array([1,2,3])
print(arr.dtype)

arr = arr.astype(np.float64)
print(arr.dtype)
```

```python
#dictionary

dc = {'a': 10, 'b': 20, 'c' : 30}

print(type(dc))

keys = dc.keys()
values = dc.values()
print(values)
items = dc.items()
print(items)
```

```python
#keys to ndarray
arr = np.array(list(keys))
print(type(arr))
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#values to ndarray
arr = np.array(list(values))
print(type(arr))
print(arr)
arr = arr.astype(np.int32)
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#items to ndarray
arr = np.array(list(items))
print(type(arr))
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

# Creating ndarray from scratch

```python
arr = np.zeros((2,3)) # shape (2,3) tuple
arr = arr.astype(np.int64)
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#zeros like

arr = np.zeros_like(arr3)
print(arr)
print(arr.shape)
```

```python
#np ones
arr = np.ones((4,3)) # shape (4,3) tuple
arr = arr.astype(np.int64)
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#ones like
arr = np.ones_like(arr3)
print(arr)
print(arr.shape)
```

```python
#np empty
arr = np.empty((4,3)) # shape (4,3) tuple
arr = arr.astype(np.int64)
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#empty like
arr = np.empty_like(arr3)
print(arr)
print(arr.shape)
```

```python
#np full
arr = np.full((4,3),fill_value=10) # shape (4,3) tuple
arr = arr.astype(np.int64)
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#empty like
arr = np.full_like(arr3, np.inf,dtype=np.float64)
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

# Array creation from random values
