# Class and Objects

```python
class Phone:
  category = "Electronics"
  #constructor


  def __init__(self,model, battery, camera,battery_parcentage=100):
    self.model = model
    self.battery = battery
    self.camera = camera
    self.battery_parcentage = battery_parcentage

  def charge(self, hour):
    self.battery_parcentage += hour
```

```python
## creating
apple = Phone("IPhone 15", 3000, 40)
apple.category = "Phone for Dumb People"
blueberry = Phone("B-17", 4000,30)
motorolla = Phone("M-17",3500,25)
print(apple.category)
print(blueberry.category)
print(motorolla.category)
```

```python
print(apple.camera)
print(motorolla.camera)
```

Single Inheritance

```python
class SmartPhone(Phone):
  def __init__(self,model, battery, camera,processor):
    super().__init__(model,battery,camera)
    self.processor = processor

  def charge(self, hour):
    print("fast charging in process")
    super().charge(hour)
```

```python
pro =SmartPhone("X",5000,100,"snapdragon")

print(pro.model)
pro.charge(1)
print(pro.battery_parcentage)
```

Multiple Inheritance

```python
class Cooling_Machanism:
  def __init__(self,cooling_method):
    self.cooling_method = cooling_method

  def cooling_on(self):
    print('Cooling on')
```

```python
class SmartPhoneWithCooling(SmartPhone,Cooling_Machanism):
  def __init__(self, model, battery, camera, processor,cooling_method):
    SmartPhone.__init__(self, model, battery, camera, processor)
    Cooling_Machanism.__init__(self,cooling_method)
```

```python
pro_cooling = SmartPhoneWithCooling("X",5000,100,"snapdragon","water")
print()
```

```python
class Phone_encap:
  def __init__(self,name,model,imei):
    self.name = name
    self.model = model
    self.__imei = imei

  def get_imei(self):
    return self.__imei

  def set_imei(self,imei):
    self.__imei = imei

```

abstraction


```python
from abc import ABC, abstractmethod

class Telephone(ABC):
  def make_call(self):
    print('making a call')

  @abstractmethod
  def capture_photo(self):
    pass
  @abstractmethod
  def record_video(self):
    pass
```

```python
class SPhone(Telephone):
  # abstract method implement na korle eita error dibe
  def capture_photo(self):
    print("capture photo")
  def record_video(self):
    print("record video")
```

```python
sp = SPhone()
sp.make_call()
sp.capture_photo()
sp.record_video()
```
