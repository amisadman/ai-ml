# Read File

```python
file = open("./sample_data/sample.txt","r")
#readfile() = fine in a string
#readfilelines() = every line in a list of lines
content = file.readlines()
content = list(map(str.strip, content))
print(content)
print(file.closed)
file.close()
print(file.closed)
```

```python
#with use korle auto close hoi
with open("./sample_data/sample.txt","r") as file:
  content = file.readlines()
  print(content)

print(file.closed)
```

# Write File

```python
#creating and writing on a file
with open("./sample_data/test.txt","w") as file:
  file.write("Hello test file\n")
  file.write("How is learning going on\n")
```

```python
#writing on a existing file
#override
with open("./sample_data/test.txt","w") as file:
  file.write("second attempt\n")
  file.write("learning is going fast\n")
```

```python
#no override
#append mode
with open ("./sample_data/test.txt","a") as file:
  file.write("third attempt\n")
  file.write("learning is going fast\n")
```

# Pointer tell() seek()

```python
with open ("./sample_data/sample.txt","r") as file:
  print(file.tell())
  print(file.read(5))
  print(file.tell())
  print(file.read())
  print(file.tell())
```

```python
with open ("./sample_data/sample.txt","r") as file:
  print(file.tell())
  print(file.read(5))
  print(file.tell())
  file.seek(0) # seek er moddhe index pass kori
  print(file.tell())

  #read vs seek
  #read -> samne jay
  #seek -> samne pichone dui jaygay e jay
```

```python
from functools import reduce
with open("./sample_data/simple.txt") as file:
  lines = file.readlines()
  print(len(lines))
  words = list(map(lambda x : len(x.split()),lines))
  print(sum(words))
  num_of_words = reduce(lambda x,y : x+y,words)
  print(num_of_words)
```

# Exception Handling

```python
n = int(input())

try:
  a = n/0
except ZeroDivisionError:
  print("Cannot divide by zero")
```

```python
n = int(input())

try:
  a = n/0
except Exception as e:
  print(e)
```
