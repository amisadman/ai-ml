```python
import torch
import torch.nn as nn

image_features = torch.tensor([[0.15, 0.85, 0.25, 0.70]])

model = nn.Sequential(
    nn.Linear(4,12),
    nn.ReLU(),
    nn.Linear(12,1),
    nn.Sigmoid()
)


with torch.no_grad():
    prediction = model(image_features)

prob = prediction.item()
print(f"Probability of being a Dog: {prob:.4f}")


if prob == 0.5:
   print("Couldn't classify the image")

elif prob > 0.5:
  print("Dog")
else:
  print("Cat")
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

data = load_breast_cancer()
X, y = data.data, data.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

k_values = range(1, 51)
accuracies = []

print(f"{'K Value'} | {'Test Accuracy'}")


for k in k_values:
    # Initialize and fit
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    # Calculate accuracy
    score = knn.score(X_test, y_test)
    accuracies.append(score)


print(accuracies)

print(len(accuracies))


mx_accuracy = max(accuracies)

print(mx_accuracy)

best_k_index = accuracies.index(mx_accuracy)

best_k = k_values[best_k_index]


print(best_k)

```

```python
#generator
lst = [x for x in range(500)] #data
def data_loader(chunk_size,lst):
  for i in range(0,len(lst),chunk_size):
    yield lst[i: i+chunk_size]


x = data_loader(10,lst)
print(type(x))
print(x)
print(next(x))
print(next(x))
print(next(x))
print(next(x))
```
