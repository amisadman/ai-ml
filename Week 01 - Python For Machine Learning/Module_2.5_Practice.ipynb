```python
def introduce(name, age):
  print(f"My name is {name} and I am {age} years old.")
```

```python
introduce("Rahim", 28)
```

```python
def batch_generator(data,batch_size):
  for i in range(0, len(data),batch_size):
    yield data[i: i+batch_size]
```

```python
data = [1, 2, 3, 4, 5, 6, 7, 8]
batch_size = 3

for batch in batch_generator(data,batch_size):
  print(batch)
```

```python
def square_generator():
  a = [1,2,3,4]
  squared = list(map(lambda x: x**2,a))
  print(squared)

square_generator()
```

```python
def even_square():
  a = [1,2,3,4,5,6]
  squared = list(map(lambda x : x**2,list(filter(lambda x: x%2 == 0,a))))
  print(squared)
```

```python
even_square()
```

```python
def power(base,exponent = 2):
  return base ** exponent
```

```python
print(power(5))
power(5,3)
```
